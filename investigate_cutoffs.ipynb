### Transform to add column of minimum median sell price to remove ones that are too low
If the sell price is too low there's no room for the price to drop and for anomalies to be detected

```python
csdf['min_price'] = csdf.groupby('item_name')['median_sell_price'].transform('min')
```

## I'm looking for cutoff threshold for minimum sell price still giving useful anomaly detection. Low costing items will have higher sell quantity, so they might be more resistent to anomalies, which would mean when anomalies show up they're much more likely to be real

### Find a Minimal Wear item with low sell price

```python
csdf_low_thresh = csdf[(csdf.min_price >= .04) & (csdf.min_price <= .1)]
display(csdf_low_thresh.describe())
csdf_low_thresh[csdf_low_thresh.item_name.str.contains('(Minimal Wear)')].head()
```

### Anomaly detection on low cost item

```python
plot_bizon = csdf_low_thresh[csdf_low_thresh.item_name == 'PP-Bizon | Photic Zone (Minimal Wear)'].iloc[100:]
plot_bizon = plot_bizon[['date', 'median_sell_price']].reset_index(drop=True)
plot_bizon.columns = ['timestamp', 'count']
plot_bizon.timestamp = [np.datetime64(datetime.datetime.fromtimestamp(t).date()) for t in plot_bizon.timestamp]
results = detect_ts(plot_bizon, max_anoms=8/len(plot_bizon), alpha=0.001, direction='both', only_last=None, longterm=True, verbose=True, piecewise_median_period_weeks=40)

# format the twitter data nicely
plot_bizon['timestamp'] = pd.to_datetime(plot_bizon['timestamp'])
# make a nice plot
fig, ax = plt.subplots(figsize=(22, 8))
fig.autofmt_xdate(rotation=45)
ax.plot(plot_bizon['timestamp'], plot_bizon['value'], 'b')
ax.plot(results['anoms'].index, results['anoms']['anoms'], 'ro')
ax.set_title('Detected Anomalies')
ax.set_ylabel('value')
```

```python
results['anoms']
```
